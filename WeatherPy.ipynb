```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# OpenWeatherMap API Key
from wusiapikeys import openweather
api_key = openweather["key"]

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
np.random.seed(30) #set seeds to get constant results
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    result= citipy.nearest_city(lat_lng[0], lat_lng[1])
    city = result.city_name
    country_code = result.country_code #add country code to get unambiguous result from API call
    
    # If the city,country_code is unique, then add it to a our cities list
    if (city,country_code) not in cities:
        cities.append((city,country_code))
        
# Print the city count to confirm sufficient count
len(cities)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=lats,y=lngs,c=(0.7,0.7,0.7))
plt.ylim(-190,190)
plt.xlim(-95,95)
plt.title("Random generated coordinates")
plt.show()
```

## Perform API Calls

```python
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

weather=[] # initialize list to store weather data for each city
count=1
group=1
# perform API calls for all cities in the cities list, get the current weather in Fahrenheit
for city in cities:
    iurl=url+"&q="+city[0]+","+city[1]
    result = requests.get(iurl).json()
    print(f"processing record {count} of group {group}: {city}-----------------------------------------")
    print(iurl)
    try:
        temp=result["main"]["temp"] # Temperature (F)
        humid=result["main"]["humidity"] # Humidity (%)
        cloud=result["clouds"]["all"] # Cloudiness (%)
        wind_speed=result["wind"]["speed"] # Wind Speed (mph)
        lat=result["coord"]["lat"]
        lon=result["coord"]["lon"]

        weather.append({"city":city[0],
                        "country":city[1],
                        "Temperature (F)":temp,
                        "Humidity (%)":humid,
                        "Cloudiness (%)":cloud,
                        "Wind Speed (mph)":wind_speed,
                        "Latitude":lat,
                        "Longitude":lon})
        
        print(f"successful")
    except Exception as e:
        print(f"encounter error: {e}")
    
    count+=1
    
    if count == 101:
        time.sleep(1)
        group+=1
        count=1
```

```python
# convert weather results into DataFrame
weather_df = pd.DataFrame(weather)
weather_df.head(10)
```

```python
weather_df.count()
```

```python
# export result to csv file
weather_df.to_csv(output_data_file) 
```

# Create Plots

```python
# build a series of scatter plots to showcase the following relationships:
# Temperature (F) vs. Latitude
# Humidity (%) vs. Latitude
# Cloudiness (%) vs. Latitude
# Wind Speed (mph) vs. Latitude
current_time=time.strftime('%Y/%m/%d %H:%M',time.localtime())

plt.style.use("ggplot")

for i in ("Temperature (F)","Humidity (%)","Cloudiness (%)","Wind Speed (mph)"):
    # make plot
    weather_df.plot(kind="scatter",x="Latitude",y=i,figsize=(10,6),style="o", s=40,edgecolors="b",alpha=0.5)
    plt.title(f"{i} vs. Latitude ({current_time})")
    plt.xlim(-95,95)
    
    # save plot as png file
    plt.savefig(f"output_data/{i} vs. Latitute.png")
    
    
```

#### 1. Temperature increases when it approaches the equator
#### 2. There is less chance of having a low humidity when getting closer to the equateor
#### 3. No significant relationship identified between cloudness and latitude
#### 4. No significant relationship identified between wind speed and latitude
